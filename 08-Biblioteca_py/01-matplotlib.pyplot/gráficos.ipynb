```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random sales data
vendas = np.random.randint(1000, 3000, 12)

# Generating month numbers
meses = np.arange(1, 13)  # Corrected to 11 to match the length of 'vendas'

# Create a figure with a specific size
plt.figure(1, figsize=(15, 3))

# First subplot: Line plot
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, color='red')
plt.title('Line Plot')
plt.xlabel('Month')
plt.ylabel('Sales')

# Second subplot: Bar plot with blue bars
plt.subplot(1, 3, 2)
plt.bar(meses, vendas, color='blue')
plt.title('Bar Plot (Blue)')
plt.xlabel('Month')
plt.ylabel('Sales')

# Third subplot: Bar plot with black bars
plt.subplot(1, 3, 3)
plt.bar(meses, vendas, color='black')
plt.title('Bar Plot (Black)')
plt.xlabel('Month')
plt.ylabel('Sales')

# Display the figure with the subplots
plt.tight_layout()
plt.show()
```
